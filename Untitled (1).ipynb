```python
#plot the sutabile graph to display average price vs place
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Hotels.csv')
average_price = df.groupby('name')['price'].mean()
average_price.plot(kind='bar')
plt.xlabel('place')
plt.ylabel('Average Price')
plt.title('Average Price of place')
plt.show()
```

```python
#plot the bar chart for days vs total
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Hotels.csv")
total_by_days = df.groupby('days')['total'].sum()
total_by_days.plot(kind='bar')
plt.xlabel('Number of Days')
plt.ylabel('Total Price')
plt.title('Total Price by Number of Days')
plt.show()
```
